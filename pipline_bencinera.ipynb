```python
import os
import logging
import pandas as pd

def extraer_db_historico(self):
        """leer dataset de transacciones."""
        logging.info("Extrayendo transacciones reales...")
        try:
            df = pd.read_csv("df_bencinera_final.csv")
            return df
        except Exception as e:
            logging.critical(f"Falla al leer el archivo CSV: {e}")
            return pd.DataFrame()

def transformar_datos(self, df_crudo):
        logging.info("iniciando")

        cols_interes = ['id_vehiculo', 'industria', 'cantidad', 'codigo_error', 'duracion_carga_seg']
        df_limpio = df_crudo[[c for c in cols_interes if c in df_crudo.columns]].copy()
        try:
            df_limpio['cantidad'] = pd.to_numeric(df_limpio['cantidad'], errors='coerce')
            df_limpio = df_limpio.dropna(subset=['cantidad'])
            logging.info(f"Limpieza completada: {len(df_limpio)} filas válidas.")
            return df_limpio
        except Exception as e:
            logging.error(f"Error en transformación: {e}")
            return pd.DataFrame()
```
